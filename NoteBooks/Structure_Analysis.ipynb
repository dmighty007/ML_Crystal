```python
import MDAnalysis as mda
from MDAnalysis.analysis.rdf import InterRDF
import matplotlib.pyplot as plt

dump_file = "./Cooling_Trajs/dump100_T_0.45.atom"
u = mda.Universe(dump_file, format="LAMMPSDUMP")

# all atoms
atoms = u.select_atoms("type 1")

rdf = InterRDF(
    atoms,
    atoms,
    range=(0.0, 5.0),   # r_min, r_max
    nbins=75,
    exclusion_block=(1, 1)
)

rdf.run(verbose=True, step = 1)

# r values and g(r)
r = rdf.bins
g_r = rdf.rdf

plt.plot(r, g_r)
plt.xlabel("r")
plt.ylabel("g(r)")
plt.tight_layout()
plt.show()
```
